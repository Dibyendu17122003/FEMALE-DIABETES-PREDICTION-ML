```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.metrics import accuracy_score
df = pd.read_csv("diabetes.csv")  
X = df.drop(columns="Outcome")
y = df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# MAIN SVM MODEL
classifier = SVC(kernel='linear')
classifier.fit(X_train_scaled, y_train)

X_train_Prediction = classifier.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_Prediction, y_train)
print("SVM Training Accuracy:", round(training_data_accuracy*100, 2))

X_test_Prediction = classifier.predict(X_test_scaled)
testing_data_accuracy = accuracy_score(X_test_Prediction, y_test)
print("SVM Test Accuracy:", round(testing_data_accuracy*100, 2))
# STACKING CLASSIFIER
classifier = StackingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000)),
        ('rf', RandomForestClassifier(n_estimators=200, random_state=42)),
        ('gb', GradientBoostingClassifier()),
        ('svm', SVC(kernel='linear', probability=True))
    ],
    final_estimator=LogisticRegression(max_iter=1000)
)
classifier.fit(X_train_scaled, y_train)

X_train_Prediction = classifier.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_Prediction, y_train)
print("Stacking Ensemble Training Accuracy:", round(training_data_accuracy*100, 2))

X_test_Prediction = classifier.predict(X_test_scaled)
testing_data_accuracy = accuracy_score(X_test_Prediction, y_test)
print("Stacking Ensemble Test Accuracy:", round(testing_data_accuracy*100, 2))
```
